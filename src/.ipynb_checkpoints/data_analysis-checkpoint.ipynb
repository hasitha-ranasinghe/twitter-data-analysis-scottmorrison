```python
import pandas as pd     
import numpy as np  
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from textblob import Blobber
from textblob.sentiments import NaiveBayesAnalyzer
```

```python
data = pd.read_csv('data/processed/processed_data.csv')
```

```python
data.head(20)
```

```python
blobber = Blobber(analyzer=NaiveBayesAnalyzer())

def sent(text):
    blob = blobber(text)
    return blob.sentiment.classification
```

```python
data['sentiment_classification'] = np.array([ sent(str(tweet)) for tweet in data['tweet_text']])
```

```python
data.head()
```

```python
sns.set_style('whitegrid')
sns.set_palette("GnBu_d")
```

```python
sns.countplot(x='sentiment_classification',data=data, )
```

```python
pos_tweets= [ tweet for index, tweet in enumerate(data['sentiment_classification']) if data['sentiment_classification'][index] == 'pos']
neg_tweets= [ tweet for index, tweet in enumerate(data['sentiment_classification']) if data['sentiment_classification'][index] == 'neg']

pos_percentage = len(pos_tweets)*100/len(data)
neg_percentage = len(neg_tweets)*100/len(data)
```

```python
print(pos_percentage)
print(neg_percentage)
```

```python
sns.set_palette("Paired")
```

```python
fig = plt.figure()
percentage = pd.Series([pos_percentage,neg_percentage], 
                      index=['pos tweets','neg tweets'], 
                      name='Sentimental analysis')
percentage.plot.pie(fontsize=9, autopct='%.2f%%', figsize=(6.5, 6.5));
plt.figure()
#fig.savefig('Hasitha_Graphs/figure2.png', bbox_inches='tight')
```
